```python
import pandas as pd

# 1. Load the CSV data
df = pd.read_csv('Fertilizer Guide Book.csv')
print("Original Columns:\n", df.columns.tolist())
df.head()
```

```python
# 2. Remove specified columns: Gypsum, Zn Test Value, Zn Sulphate Hepta Hydrate, Equivalent Gypsum
columns_to_drop = ['Gypsum', 'Zn Test Value', 'Zn Sulphate Hepta Hydrate', 'Equivalent Gypsum']
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')

print("Columns after dropping specified columns:\n", df_cleaned.columns.tolist())
df_cleaned.head()
```

```python
# 3. Rename columns to N, P, K
# Depending on your use case, N, P, K can refer to either the Fertilizer amounts or the Soil Test Values.

# Interpretation A: N, P, K refer to the Fertilizers (Urea = N, TSP/DAP Olsen = P, MoP = K)
df_fertilizers = df_cleaned.rename(columns={
    'Urea': 'N',
    'TSP/DAP Olsen Method': 'P',
    'MoP': 'K'
})
# Drop the redundant Bray & Kurtz P column
df_fertilizers = df_fertilizers.drop(columns=['TSP/DAP Bray & Kurtz'], errors='ignore')

print("Interpretation A (Fertilizers as N, P, K) columns:\n", df_fertilizers.columns.tolist())
df_fertilizers.head()
```

```python
# Interpretation B: N, P, K refer to the Soil Test Values
# Rename 'N/K Test Value' to 'N', 'P/S Test Value' to 'P', and construct 'K' (typically set to 'N' as K is combined in N/K)
df_soil = df_cleaned.copy()
df_soil = df_soil.rename(columns={
    'N/K Test Value': 'N',
    'P/S Test Value': 'P'
})
df_soil['K'] = df_soil['N']  # Copying N to K as combined N/K test value is used

# Reorder columns for convenience
df_soil = df_soil[['Crop', 'N', 'P', 'K', 'Urea', 'MoP', 'TSP/DAP Olsen Method', 'TSP/DAP Bray & Kurtz']]

print("Interpretation B (Soil Test Values as N, P, K) columns:\n", df_soil.columns.tolist())
df_soil.head()
```

```python
# 4. Place the P/S Test Value column directly beside the N/K Test Value column
df_reordered = df_cleaned[[
    'Crop',
    'N/K Test Value',
    'P/S Test Value',
    'Urea',
    'MoP',
    'TSP/DAP Olsen Method',
    'TSP/DAP Bray & Kurtz'
]]

print("Columns after placing P beside N/K:\n", df_reordered.columns.tolist())
df_reordered.head()
```
